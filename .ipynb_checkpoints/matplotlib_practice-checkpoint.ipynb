```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
# if we give in an arry matplotlib will read it as the y axis and 
# will generate an x-axis for you 
savings = [0, 100, 150, 200, 300, 500, 300, 350, 500, 600, 750, 1000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','July', 'Aug', 'Sep',
         'Oct', 'Nov', 'Dec']
```

```python
# This method is the matlab method
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot (Matlab Method)
plt.plot(x_axis, y_axis)
```

```python
# Creating a Chart using the Object-Oriented Method
# we will be using ax.plot() instead of plt.plot()
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# plt.subplots() returns a tuple that contains a figure and axes object
# fig, ax = plt.subplots() unpacks the tubels into variables fig and ax
# fig refers the figure, ax refers to axes of figure
# ax.plot(x_axis, y_axis) plots data from x and y axes
```

```python
# fig, ax = plt.subplots() can be written like below:
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# for a quick graph we can use ax = plt.axes(), ax.plot(x,y)
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# plt.show() looks and shows all of the active plots 
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# practicing matplotlib labeling
#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston')
#Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fares($)')
#set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title('PyBer Fare by Month')
# Add a legend
plt.legend()
```

```python
# add colors with color =, width changes width, marker = marker
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = 'Boston')
#Create labels for x and y axes
plt.xlabel('Date')
plt.ylabel('Fares ($)')
# set y-axis limit from 0-45
plt.ylim(0, 45)
plt.title('PyBer Fare by Month')
# Add grid
plt.grid()
# add legend
plt.legend()
```

```python
# to get limited range on x-axis use plt.xlim(0,5) or plt.xlim("Jan", "June")
# to get fig size change it is (w,h) where w is width and h is height
# plt.figure(figsize = (20,10)) or plt.subplots(figsize= (20,10))
```

```python
plt.plot(x_axis, y_axis, marker = 'D', color = 'green', linewidth = 2, label = 'Boston')
plt.xlabel('Date')
plt.ylabel('Fares ($)')
plt.grid()
plt.legend()
plt.ylim(0,45)
plt.title('Pyber Fare by Month')
```

```python
# creating bar charts with matplotlib using MATLAB
#Set the x-axis to list string for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create bar chart using plt.bar()
plt.bar(x_axis, y_axis)
```

```python
# we can use similar editing for the bar graphs
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# we can use a horizontal bar graph with plt.barh(x,y)
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# to flip the axes just swap them in the script
plt.barh(y_axis, x_axis)
```

```python
# we can invert the dates so that Janurary is on the top and Dec is on bot
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# lets change the color and add labels 
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.legend()
plt.gca().invert_yaxis()
```

```python
# creating line charts with object-oriented method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = 'c', label = 'Chicago')
plt.legend()
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('Pyber Fare by Month')
plt.gca().invert_yaxis()
```

```python
# create a scatter plot using plt.scatter. note: the x and y matches in index. x[0] goes with y[0]
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
```

```python
#create the scatter plot and label chart. c is for color, marker changes marker, s changes size of markers. set to diff size 
#but has to be the same number of elements in our list. 
plt.scatter(x,y, c = 'green', marker = 'x', s =x)
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Average Fare Vs Total Rides')
plt.show()
```
