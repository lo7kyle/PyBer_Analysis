```python
import matplotlib.pyplot as plt
```

```python
# if we give in an arry matplotlib will read it as the y axis and 
# will generate an x-axis for you 
savings = [0, 100, 150, 200, 300, 500, 300, 350, 500, 600, 750, 1000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','July', 'Aug', 'Sep',
         'Oct', 'Nov', 'Dec']
```

```python
# This method is the matlab method
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot (Matlab Method)
plt.plot(x_axis, y_axis)
```

```python
# Creating a Chart using the Object-Oriented Method
# we will be using ax.plot() instead of plt.plot()
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# plt.subplots() returns a tuple that contains a figure and axes object
# fig, ax = plt.subplots() unpacks the tubels into variables fig and ax
# fig refers the figure, ax refers to axes of figure
# ax.plot(x_axis, y_axis) plots data from x and y axes
```

```python
# fig, ax = plt.subplots() can be written like below:
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# for a quick graph we can use ax = plt.axes(), ax.plot(x,y)
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# plt.show() looks and shows all of the active plots 
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# practicing matplotlib labeling
#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston')
#Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fares($)')
#set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title('PyBer Fare by Month')
# Add a legend
plt.legend()
```

```python
# add colors with color =, width changes width, marker = marker
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = 'Boston')
#Create labels for x and y axes
plt.xlabel('Date')
plt.ylabel('Fares ($)')
# set y-axis limit from 0-45
plt.ylim(0, 45)
plt.title('PyBer Fare by Month')
# Add grid
plt.grid()
# add legend
plt.legend()
```

```python
# create a scatter plot using plt.scatter. note: the x and y matches in index. x[0] goes with y[0]
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
```

```python
#create the scatter plot and label chart. c is for color, marker changes marker, s changes size of markers. set to diff size 
#but has to be the same number of elements in our list. 
# Scatter plots benefits by helping depict a positive, negative, or no correlation (regression analysis)
plt.scatter(x,y, c = 'green', marker = 'x', s =x)
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Average Fare Vs Total Rides')
plt.show()
```

```python
# some markers look small so we can adjust the size by muliplying the dat in the y-axis by any value to make it look bigger
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# we can refactor the for-loop code using list comprehension. 
# syntax: new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# creating a pie chart helps visualize percentages 
categories = ['pants', 'shirts', 'outerwear', 'pajamas', 'underwear', 'shoes']
revenue =  [25, 15, 5, 20, 10, 25]
```

```python
# For the plt.pie you only need the percentages for the arg
plt.pie(revenue)
plt.show()
```

```python
# we can format and label our pie chart using label = categories and autopercent autopct for percentage formatting
plt.pie(revenue, labels = categories, autopct = '%1.1f%%')
plt.show()
```

```python
# we can use explode to highlight a specific value. we do this by using a list of values between 0,1 that is the same size
explode = [0, .1, 0, 0, 0, 0]
```

```python
# using explode method
plt.pie(revenue, labels = categories, autopct = '%1.1f%%', explode = explode)
plt.show()
```

```python
# Create pie chart with MATLAB
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
# using explode and autopct. the % before/after .1f formats number a a percentage, .1f is for 1 decimal place
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# we can change the size of the chart so the numbers wont overlap using plt.subplots(figsize = 8,8))
# similar with other plots we can change the colors using colors = '' but we need to assign 1 color for each month
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# pie charts with object-oriented method
fig, ax = plt.subplots()
ax.pie(y_axis, labels = x_axis)
plt.show()
```

```python
# format the pie chart via color and popping march and july. Start Janurary at the top using the gca().invert_yaxis()
# pie charts starts with counterclock as True, and start angle at 0
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    startangle = 90,
    counterclock = False)

plt.show()
```
