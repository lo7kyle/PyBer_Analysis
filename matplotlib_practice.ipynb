```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
# if we give in an arry matplotlib will read it as the y axis and 
# will generate an x-axis for you 
savings = [0, 100, 150, 200, 300, 500, 300, 350, 500, 600, 750, 1000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','July', 'Aug', 'Sep',
         'Oct', 'Nov', 'Dec']
```

```python
# This method is the matlab method
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot (Matlab Method)
plt.plot(x_axis, y_axis)
```

```python
# Creating a Chart using the Object-Oriented Method
# we will be using ax.plot() instead of plt.plot()
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# plt.subplots() returns a tuple that contains a figure and axes object
# fig, ax = plt.subplots() unpacks the tubels into variables fig and ax
# fig refers the figure, ax refers to axes of figure
# ax.plot(x_axis, y_axis) plots data from x and y axes
```

```python
# fig, ax = plt.subplots() can be written like below:
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# for a quick graph we can use ax = plt.axes(), ax.plot(x,y)
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# plt.show() looks and shows all of the active plots 
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# practicing matplotlib labeling
#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston')
#Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fares($)')
#set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title('PyBer Fare by Month')
# Add a legend
plt.legend()
```

```python
# add colors with color =, width changes width, marker = marker
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = 'Boston')
#Create labels for x and y axes
plt.xlabel('Date')
plt.ylabel('Fares ($)')
# set y-axis limit from 0-45
plt.ylim(0, 45)
plt.title('PyBer Fare by Month')
# Add grid
plt.grid()
# add legend
plt.legend()
```
